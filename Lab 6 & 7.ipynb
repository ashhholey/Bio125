# Laboratory 6/7 Sensory Physiology

## Purpose 

Sensation, the monitoring of  stimuli, involves the interaction of 3 basic components of the nervous system. Receptors generate impulses in response to specific environmental stimuli, sensory neuron relay these impulses through afferent pathways to the central nervous system & interpretation centers of the cerebral cortex translate these impulses into perceived sensations.

## Procedures

### A Tests of cutaneous sensation
A-1 Two point discrimination
1. with your partner's eyes closed, apply 2 caliper pinpoints as closely together as possible on your partner's skin on the palm of her hand
2. remove the pins and move them 1 millimeter apart. Reapply the caliper points to your partner's skin. repeat until partner can discriminate 2 distinct points
3. record distance between pins at which your partner can discriminate two seperate caliper points
4. compare results 
5. repeat with other partner

A-2 Accommodation of thermoreceptors
1. place your left fingers in 15 degree celcius water and your right finger in warm water 37 degree celcius and record sensation of of each
2. after 2 minutes, describe the sensation in each hand
3. remove hands and promptly place them both in 25 degree celcius water. describe the immediate sensation in each hand

### B Olfactory adaptation
1. block your left nostril. uncork and hold the bottle of camphor oil under your nose until you can no longer detect tge camphor. record adaptation time
2. remove the camphor and place the bottles of cloves, then peppermint oil under your nose. distinguish the smells of cloves and peppermint oil
3. uncork and hold the bottle of camphor under your nose again until the smell is no longer recognized. record second adaptation time
4. unblock your left nostril determine if the camphor is detected

### C Auditory measurements
C-1 tuning fork tests 
a. Rinne's test
    1. plug your left ear with cotton or hold your hand over it and test the right ear
    2. hold the handle of a vibrating tuning fork to the right mastoid process
    3. when the sound disappears, move the fork near the external auditory canal
    4. reappearance of the sound indicates no middle ear damage
    5. repeat to left ear
    6. record results for each ear

### D Equilibrium
1. a student volunteer will be seated on a swivel stool with her head bent at 30 degrees forward
2. the instructor will spin the student rapidly for 10 turns
3. the instructor will suddently stop turning the student and have the student look straight ahead
4. observe and note the subsequent movent of the student's eyes

### E Visual Measurements
E-1 demonstration of the blind spot
1. cover your left eye and focus the right eye on the center of the cross in lab manual
2. slowly bring the page closer to your eye until the spot disappears
3. have your partner measure this distance from your eye to the page
4. the image of the spot is now superimposed on the optic nerve

E-2 The Snellen test
1. stand 20 ft away from the Snellen chart and cover left eye
2. attempt to read the line designated "20"
3. if you cannot read line 20, attempt 30, 40, 50, 70, 100, or 200 until a line is legible. perform these attempts with your left eye, covering the right eye
4. the Snellen chart is analyzed on pg 39 of lab manual

E-3 Astigmatism
1. stand approximately 8-10 inches away from the radial astigmatism eye chart so that it fills your field of vision, cover left eye
2. focus on the lines in the vertical plane with your right eye
3. if a blur sppears in the lateral lines or the lines converge into one, you have an astigmatism in this plane of your eye
4. record results of this test and repeat on other eye

E-4 Color vision: Color blind-ness test
1. ontain the Ichikawa color blindness charts
2. attempt to read the numbers of each pattern on the test panels
3. after the first 10 test panels, if your score indicates color blindness, continue with the next five test panels to determine which color deficiency exists

E-5 Perimetry 
1. seat yourself before the perimeter board with your right eye at the edge of the semicircle. cover your left eye. stare at the center line
2. your lab partner will introduce several different colored blocks into your field of vision. identify these blocks by color. do not take your eye from the center of the chart or uncover your left eye
3. your partner will record the degree at which the colors were discriminated on the perimetry score sheet on pg 47 of the lab manual
4. repeat these procedures for each block for both the horizontal and vertical perimetry charts. record the data and connect the same colored dots to form an outline of cone placement of your right eye on your data sheet
